Find the director who made most movies?

```python
#Fetching the director who made most movies
cur.execute("SELECT d.name FROM MOVIES AS m JOIN directors AS d ON m.director_id == d.id GROUP BY m.director_id ORDER BY COUNT(m.title) DESC LIMIT 1")
most_movies = cur.fetchall()
# Displaying the director who made most movies
print(f"The director who made the most movies is '{most_movies[0][0]}'")
```

Name of the director who is most bankable?

```python
# Fetching the name of the director who is most bankable
cur.execute("SELECT d.name FROM movies as m JOIN directors as d ON m.director_id == d.id GROUP BY director_id ORDER BY SUM(m.budget) DESC LIMIT 1")
most_bankable = cur.fetchall()
# Displaying the name of the director who is most bankable
print(f"The most bankable director is {most_bankable[0][0]}")
```

Director Analysis

Find all the directors with the number of movies and revenue?

```python
# Fetching all the directors with the number of movies and revenue
cur.execute("SELECT d.name, COUNT(*) AS 'Total Movies',SUM(revenue) AS revenue FROM movies AS m JOIN directors AS d ON m.director_id == d.id GROUP BY director_id")
director_revenue = cur.fetchall()
# Displaying the data frame of all the directors with the number of movies and revenue
director_most_revenue = pd.DataFrame(director_revenue,columns=['Director Name','Total Movies','revenue'])
director_most_revenue
```

Find all the directors with the number of movies and revenue where the no. of movies should be taken into account for doing analysis.The director who has highest no. of movies should comes at the top and so on and so forth?

```python
# Fetching the director who has highest no. of movies should comes at the top
cur.execute("SELECT d.name, COUNT(m.title) AS 'Total Movies', SUM(m.revenue) AS revenue FROM movies AS m JOIN directors AS d ON m.director_id == d.id GROUP BY director_id ORDER BY COUNT(m.title) DESC LIMIT 10")
director_revenue = cur.fetchall()
# Displaying the data frame of the director who has highest no. of movies should comes at the top
director_most_revenue = pd.DataFrame(director_revenue,columns=['Director Name','Total Movies','revenue'])
director_most_revenue
```
